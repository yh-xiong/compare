```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import numpy as np
import os
from tqdm import tqdm
import pickle
from typing import Tuple, List, Optional, Dict
from torch.optim.lr_scheduler import CosineAnnealingLR, OneCycleLR, ReduceLROnPlateau
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, f1_score, classification_report, confusion_matrix
from sklearn.metrics import normalized_mutual_info_score, recall_score, precision_score
import seaborn as sns
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')


# ==================== 1. 数据增强策略 ====================
class AdvancedSimCLRTransform:
    """增强的数据增强策略"""
    def __init__(self, image_size=96, strong_aug=True):
        self.image_size = image_size
        self.strong_aug = strong_aug
        
        # STL-10标准化参数
        self.mean = [0.4409, 0.4279, 0.3870]
        self.std = [0.2683, 0.2611, 0.2687]
        
    def __call__(self, x):
        if self.strong_aug:
            # 第一个增强视图
            transform1 = transforms.Compose([
                transforms.RandomResizedCrop(size=self.image_size, scale=(0.08, 1.0)),
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.RandomApply([
                    transforms.ColorJitter(brightness=0.8, contrast=0.8, saturation=0.8, hue=0.2)
                ], p=0.8),
                transforms.RandomGrayscale(p=0.2),
                transforms.RandomApply([transforms.GaussianBlur(kernel_size=3)], p=0.5),
                transforms.ToTensor(),
                transforms.Normalize(mean=self.mean, std=self.std),
            ])
            
            # 第二个增强视图
            transform2 = transforms.Compose([
                transforms.RandomResizedCrop(size=self.image_size, scale=(0.08, 1.0)),
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.RandomApply([
                    transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1)
                ], p=0.8),
                transforms.RandomGrayscale(p=0.2),
                transforms.RandomApply([transforms.GaussianBlur(kernel_size=3)], p=0.5),
                transforms.ToTensor(),
                transforms.Normalize(mean=self.mean, std=self.std),
            ])
        else:
            # 弱增强
            transform1 = transforms.Compose([
                transforms.RandomResizedCrop(size=self.image_size, scale=(0.2, 1.0)),
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.ToTensor(),
                transforms.Normalize(mean=self.mean, std=self.std),
            ])
            transform2 = transform1
        
        return transform1(x), transform2(x)


# ==================== 2. 模型架构 ====================
class ImprovedResNetBackbone(nn.Module):
    """改进的ResNet骨干网络"""
    def __init__(self, base_encoder='resnet50', pretrained=False):
        super().__init__()
        
        # 选择backbone
        if base_encoder == 'resnet18':
            encoder = torchvision.models.resnet18
            self.global_dim = 512
            self.local_dim = 256
        elif base_encoder == 'resnet34':
            encoder = torchvision.models.resnet34
            self.global_dim = 512
            self.local_dim = 256
        elif base_encoder == 'resnet50':
            encoder = torchvision.models.resnet50
            self.global_dim = 2048
            self.local_dim = 1024
        else:
            raise ValueError(f"Unsupported base_encoder: {base_encoder}")
        
        # 加载模型
        try:
            if pretrained:
                self.backbone = encoder(weights="DEFAULT")
            else:
                self.backbone = encoder(weights=None)
        except:
            if pretrained:
                self.backbone = encoder(pretrained=True)
            else:
                self.backbone = encoder(pretrained=False)
        
        # 移除分类头
        self.backbone.fc = nn.Identity()
        
        # 提取中间层
        self.layer1 = self.backbone.layer1
        self.layer2 = self.backbone.layer2
        self.layer3 = self.backbone.layer3
        self.layer4 = self.backbone.layer4
    
    def forward(self, x):
        # 提取特征
        x = self.backbone.conv1(x)
        x = self.backbone.bn1(x)
        x = self.backbone.relu(x)
        x = self.backbone.maxpool(x)
        
        x = self.layer1(x)
        x = self.layer2(x)
        x3 = self.layer3(x)  # layer3，用于局部特征
        x4 = self.layer4(x3)  # layer4，用于全局特征
        
        # 全局特征
        x_global = nn.functional.adaptive_avg_pool2d(x4, (1, 1))
        x_global = x_global.flatten(1)
        
        # 局部特征
        x_local = nn.functional.adaptive_avg_pool2d(x3, (4, 4))
        
        return x_global, x_local


class ImprovedProjectionHead(nn.Module):
    """改进的投影头"""
    def __init__(self, input_dim, output_dim=256, hidden_dim=2048):
        super().__init__()
        
        self.mlp = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU(inplace=True),
            nn.Dropout(0.1),
            nn.Linear(hidden_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU(inplace=True),
            nn.Linear(hidden_dim, output_dim),
        )
    
    def forward(self, x):
        return self.mlp(x)


class ImprovedSimCLRModel(nn.Module):
    """SimCLR模型"""
    def __init__(self, backbone, global_projector):
        super().__init__()
        self.encoder = backbone
        self.projector = global_projector
    
    def forward(self, x, return_features=False):
        # 提取特征
        global_features, _ = self.encoder(x)
        
        # 投影特征
        projections = self.projector(global_features)
        projections = nn.functional.normalize(projections, dim=1)
        
        if return_features:
            return projections, global_features
        return projections, global_features


# ==================== 3. 损失 ====================
def improved_simclr_loss(z1, z2, temperature=0.07):
    batch_size = z1.size(0)
    
    # 归一化
    z1 = nn.functional.normalize(z1, dim=1)
    z2 = nn.functional.normalize(z2, dim=1)
    
    # 构建相似度矩阵
    representations = torch.cat([z1, z2], dim=0)  # [2B, D]

    # 1. 计算余弦相似度
    cos_sim = torch.matmul(representations, representations.T)
    cos_sim = torch.clamp(cos_sim, max=0.95)
    similarity_matrix = cos_sim / temperature
    
    # 创建标签：每个样本的正例是它的增强版本
    labels = torch.cat([torch.arange(batch_size) for _ in range(2)], dim=0)
    labels = (labels.unsqueeze(0) == labels.unsqueeze(1)).float()
    labels = labels.to(z1.device)
    
    # 排除自对比（对角线）
    mask = torch.eye(2 * batch_size, dtype=torch.bool, device=z1.device)
    labels = labels.masked_fill(mask, 0.0)
    
    # 计算softmax
    logits_max, _ = torch.max(similarity_matrix, dim=1, keepdim=True)
    logits = similarity_matrix - logits_max.detach()  # 稳定化
    
    exp_logits = torch.exp(logits)
    log_prob = logits - torch.log(exp_logits.sum(1, keepdim=True))
    
    # 计算平均负对数似然
    mean_log_prob_pos = (labels * log_prob).sum(1) / labels.sum(1)
    loss = -mean_log_prob_pos.mean()
    
    return loss


# ==================== 4. 训练循环 ====================
def train_epoch_improved(model, train_loader, optimizer, device, epoch,
                        temperature=0.07, grad_clip=1.0, use_amp=True):
    """训练epoch"""
    model.train()
    total_loss = 0
    num_batches = len(train_loader)
    
    # 混合精度训练的scaler
    scaler = torch.cuda.amp.GradScaler() if use_amp else None
    
    pbar = tqdm(train_loader, desc=f"Epoch {epoch}")
    for batch_idx, (v1, v2, _) in enumerate(pbar):
        v1, v2 = v1.to(device, non_blocking=True), v2.to(device, non_blocking=True)
        
        optimizer.zero_grad()
        
        if scaler is not None:  # 混合精度训练
            with torch.cuda.amp.autocast():
                z1, _ = model(v1)
                z2, _ = model(v2)
                loss = improved_simclr_loss(z1, z2, temperature)
            
            scaler.scale(loss).backward()
            if grad_clip > 0:
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=grad_clip)
            scaler.step(optimizer)
            scaler.update()
        else:
            z1, _ = model(v1)
            z2, _ = model(v2)
            loss = improved_simclr_loss(z1, z2, temperature)
            
            loss.backward()
            if grad_clip > 0:
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=grad_clip)
            optimizer.step()
        
        total_loss += loss.item()
        
        # 更新进度条
        avg_loss = total_loss / (batch_idx + 1)
        pbar.set_postfix({
            'Loss': f'{avg_loss:.4f}',
            'LR': f'{optimizer.param_groups[0]["lr"]:.6f}'
        })
    
    return total_loss / num_batches


# ==================== 5. 线性评估====================
def linear_evaluation_simple(model, train_loader, test_loader, 
                            num_classes=10, epochs=100, device='cuda'):
    model.eval()
    for param in model.parameters():
        param.requires_grad = False
    
    # 简单的线性分类器
    linear_classifier = nn.Sequential(
        nn.Linear(256, 512),
        nn.BatchNorm1d(512),
        nn.ReLU(inplace=True),
        nn.Dropout(0.5),
        nn.Linear(512, 256),
        nn.BatchNorm1d(256),
        nn.ReLU(inplace=True),
        nn.Dropout(0.3),
        nn.Linear(256, num_classes)
    ).to(device)
    
    optimizer = optim.AdamW(linear_classifier.parameters(), lr=0.01, weight_decay=1e-4)
    scheduler = CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-6)
    criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
    
    best_acc = 0.0
    train_accs = []
    test_accs = []
    
    all_prec = []
    all_recall = []
    all_f1 = []
    
    for epoch in range(epochs):
        # 训练阶段
        linear_classifier.train()
        correct = 0
        total = 0
        
        pbar = tqdm(train_loader, desc=f"Linear Eval Epoch {epoch+1}/{epochs}")
        for batch_idx, (images, labels) in enumerate(pbar):
            images, labels = images.to(device), labels.to(device)
            
            optimizer.zero_grad()
            
            with torch.no_grad():
                z, _ = model(images)
            
            logits = linear_classifier(z)
            loss = criterion(logits, labels)
            loss.backward()
            optimizer.step()
            
            _, predicted = logits.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
            
            pbar.set_postfix({
                'Loss': f'{loss.item():.4f}',
                'Acc': f'{100.*correct/total:.2f}%'
            })
        
        scheduler.step()
        
        train_acc = 100. * correct / total
        train_accs.append(train_acc)
        
        # 测试阶段
        linear_classifier.eval()
        test_correct = 0
        test_total = 0
        all_predictions = []
        all_labels = []
        
        with torch.no_grad():
            for images, labels in test_loader:
                images, labels = images.to(device), labels.to(device)
                z, _ = model(images)
                logits = linear_classifier(z)
                _, predicted = logits.max(1)
                test_total += labels.size(0)
                test_correct += predicted.eq(labels).sum().item()
                all_predictions.extend(predicted.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
        
        test_acc = 100. * test_correct / test_total
        test_accs.append(test_acc)
        
        # 新增指标
        f1 = f1_score(all_labels, all_predictions, average='macro')
        prec = precision_score(all_labels, all_predictions, average='macro', zero_division=0)
        recall = recall_score(all_labels, all_predictions, average='macro', zero_division=0)
        
        all_prec.append(prec)
        all_recall.append(recall)
        all_f1.append(f1)
        
        print(f'Epoch [{epoch+1}/{epochs}], Train Acc: {train_acc:.2f}%, '
              f'Test Acc: {test_acc:.2f}%, F1: {f1:.4f}, Prec: {prec:.4f}, Recall: {recall:.4f}')
        
        # 保存最佳模型
        if test_acc > best_acc:
            best_acc = test_acc
            torch.save({
                'epoch': epoch,
                'model_state_dict': linear_classifier.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'accuracy': test_acc,
                'f1_score': f1
            }, 'checkpoints_improved/best_linear_classifier.pth')
    
    # 最终评估
    checkpoint = torch.load('checkpoints_improved/best_linear_classifier.pth')
    linear_classifier.load_state_dict(checkpoint['model_state_dict'])
    linear_classifier.eval()
    
    final_predictions = []
    final_labels = []
    
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            z, _ = model(images)
            logits = linear_classifier(z)
            _, predicted = logits.max(1)
            final_predictions.extend(predicted.cpu().numpy())
            final_labels.extend(labels.cpu().numpy())
    
    # 生成分类报告
    report = classification_report(final_labels, final_predictions, 
                                 target_names=[f'class_{i}' for i in range(num_classes)])
    
    return {
        'best_test_acc': best_acc,
        'final_test_acc': test_accs[-1],
        'final_f1_score': f1_score(final_labels, final_predictions, average='macro'),
        'final_precision': precision_score(final_labels, final_predictions, average='macro'),
        'final_recall': recall_score(final_labels, final_predictions, average='macro'),
        'classification_report': report,
        'train_acc_history': train_accs,
        'test_acc_history': test_accs,
        'precision_history': all_prec,
        'recall_history': all_recall,
        'f1_history': all_f1
    }


# ==================== 6. 特征评估 ====================
def evaluate_features_improved(model, data_loader, device, num_samples=1000):
    model.eval()
    
    all_features = []
    all_labels = []
    
    with torch.no_grad():
        for batch_idx, (images, labels) in enumerate(data_loader):
            images = images.to(device)
            
            # 提取特征
            z, _ = model(images)
            
            all_features.append(z.cpu().numpy())
            all_labels.append(labels.numpy())
            
            if len(all_features) * images.size(0) >= num_samples:
                break
    
    all_features = np.concatenate(all_features, axis=0)[:num_samples]
    all_labels = np.concatenate(all_labels, axis=0)[:num_samples]
    
    # 轮廓系数
    if len(np.unique(all_labels)) > 1:
        silhouette = silhouette_score(all_features, all_labels)
        nmi = normalized_mutual_info_score(all_labels, np.argmax(all_features, axis=1))
    else:
        silhouette = 0
        nmi = 0
    
    return {
        'silhouette_score': silhouette,
        'nmi_score': nmi,
        'num_features': len(all_features),
        'feature_dim': all_features.shape[1]
    }


# ==================== 7. 损失收敛指标计算 ====================
def compute_convergence_metrics(losses, patience=5, threshold=1e-4):
    """
    计算：
    1. 损失下降速率 Rate_L
    2. 损失波动幅度 ΔL
    3. 收敛迭代次数 T_conv
    """
    losses = np.array(losses)
    num_epochs = len(losses)
    
    # 下降速率
    rate_l = (losses[0] - losses[-1]) / num_epochs
    
    # 波动幅度
    delta_l = np.max(losses) - np.min(losses)
    
    # 收敛迭代数
    t_conv = num_epochs
    for i in range(1, num_epochs - patience):
        recent = losses[i:i+patience]
        if np.max(np.abs(np.diff(recent))) < threshold:
            t_conv = i
            break
    
    return {
        'loss_drop_rate': float(rate_l),
        'loss_fluctuation': float(delta_l),
        'convergence_epoch': t_conv
    }


# ==================== 8. 主训练函数 ====================
def main_improved():
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")
    
    if device == 'cuda':
        print(f"GPU: {torch.cuda.get_device_name(0)}")
        print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.2f} GB")
    
    # 超参数配置
    config = {
        'batch_size': 256,
        'learning_rate': 3e-4,
        'temperature': 0.07,
        'epochs_pretrain': 50,
        'grad_clip': 1.0,
        'base_encoder': 'resnet50',
        'pretrained': True,
        'use_amp': True,
        'projection_dim': 256,
        'warmup_epochs': 10,
    }
    
    os.makedirs('checkpoints_improved', exist_ok=True)
    os.makedirs('logs_improved', exist_ok=True)
    
    # 数据准备
    print("--- Preparing Data ---")
    image_size = 96
    
    class DualViewDataset:
        def __init__(self, base_dataset, transform):
            self.base_dataset = base_dataset
            self.transform = transform
            
        def __len__(self):
            return len(self.base_dataset)
            
        def __getitem__(self, idx):
            img, label = self.base_dataset[idx]
            v1, v2 = self.transform(img)
            return v1, v2, label
    
    train_dataset_raw = datasets.STL10(
        root='./data', split='unlabeled', download=True, 
        transform=None
    )
    
    simclr_transform = AdvancedSimCLRTransform(image_size=image_size, strong_aug=True)
    train_dataset = DualViewDataset(train_dataset_raw, simclr_transform)
    
    num_workers = 4
    train_loader = DataLoader(
        train_dataset, 
        batch_size=config['batch_size'], 
        shuffle=True, 
        num_workers=num_workers, 
        drop_last=True, 
        pin_memory=True
    )
    
    # 模型初始化
    print("--- Initializing Model ---")
    backbone = ImprovedResNetBackbone(
        base_encoder=config['base_encoder'],
        pretrained=config['pretrained']
    )
    
    projection_head = ImprovedProjectionHead(
        input_dim=backbone.global_dim,
        output_dim=config['projection_dim'],
        hidden_dim=2048
    )
    
    model = ImprovedSimCLRModel(backbone, projection_head).to(device)
    
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Total parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")
    
    if torch.cuda.device_count() > 1:
        print(f"Using {torch.cuda.device_count()} GPUs")
        model = nn.DataParallel(model)
    
    # 优化器
    optimizer = optim.AdamW(
        model.parameters(), 
        lr=config['learning_rate'], 
        weight_decay=1e-4,
        betas=(0.9, 0.999)
    )
    
    def warmup_cosine_scheduler(epoch, warmup_epochs=10, total_epochs=200, base_lr=3e-4):
        if epoch < warmup_epochs:
            return base_lr * (epoch + 1) / warmup_epochs
        else:
            progress = (epoch - warmup_epochs) / (total_epochs - warmup_epochs)
            return 0.5 * base_lr * (1 + np.cos(np.pi * progress))
    
    scheduler = torch.optim.lr_scheduler.LambdaLR(
        optimizer, 
        lr_lambda=lambda epoch: warmup_cosine_scheduler(
            epoch, config['warmup_epochs'], config['epochs_pretrain'], config['learning_rate']
        )
    )
    
    # 训练
    print(f"--- Starting Training for {config['epochs_pretrain']} Epochs ---")
    all_losses = []
    best_loss = float('inf')
    
    for epoch in range(config['epochs_pretrain']):
        avg_loss = train_epoch_improved(
            model, train_loader, optimizer, device, epoch+1,
            config['temperature'], config['grad_clip'], config['use_amp']
        )
        
        scheduler.step()
        all_losses.append(avg_loss)
        
        current_lr = optimizer.param_groups[0]["lr"]
        print(f'Epoch [{epoch+1}/{config["epochs_pretrain"]}]: Loss: {avg_loss:.4f}, LR: {current_lr:.6f}')
        
        if avg_loss < best_loss:
            best_loss = avg_loss
            torch.save({
                'epoch': epoch + 1,
                'model_state_dict': model.module.state_dict() if hasattr(model, 'module') else model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'loss': avg_loss,
            }, 'checkpoints_improved/best_model.pth')
    
    # 保存最终模型
    torch.save({
        'model_state_dict': model.module.state_dict() if hasattr(model, 'module') else model.state_dict(),
        'losses': all_losses,
    }, 'checkpoints_improved/final_model.pth')
    
    # ============== 新增：计算收敛指标 ==============
    conv_metrics = compute_convergence_metrics(all_losses)
    print("\n===== Convergence Metrics =====")
    print(f"损失下降速率 Rate_L: {conv_metrics['loss_drop_rate']:.6f}")
    print(f"损失波动幅度 ΔL: {conv_metrics['loss_fluctuation']:.4f}")
    print(f"收敛迭代数 T_conv: {conv_metrics['convergence_epoch']}")
    
    # 特征评估
    print("\n--- Feature Evaluation ---")
    eval_transform = transforms.Compose([
        transforms.Resize((image_size, image_size)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.4409, 0.4279, 0.3870],
                           std=[0.2683, 0.2611, 0.2687]),
    ])
    
    eval_dataset = datasets.STL10(root='./data', split='test', transform=eval_transform, download=True)
    eval_loader = DataLoader(eval_dataset, batch_size=config['batch_size'], shuffle=False, num_workers=num_workers)
    feature_metrics = evaluate_features_improved(model, eval_loader, device)
    
    print("===== Feature Quality =====")
    print(f"轮廓系数 Silhouette: {feature_metrics['silhouette_score']:.4f}")
    print(f"互信息得分 NMI: {feature_metrics['nmi_score']:.4f}")
    
    # 线性评估
    print("\n--- Linear Evaluation ---")
    linear_transform = transforms.Compose([
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomCrop(96, padding=4),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.4409, 0.4279, 0.3870], std=[0.2683, 0.2611, 0.2687]),
    ])
    
    linear_train_dataset = datasets.STL10(root='./data', split='train', transform=linear_transform, download=True)
    linear_test_dataset = datasets.STL10(root='./data', split='test', transform=eval_transform, download=True)
    
    linear_train_loader = DataLoader(linear_train_dataset, 128, shuffle=True, num_workers=num_workers)
    linear_test_loader = DataLoader(linear_test_dataset, 128, shuffle=False, num_workers=num_workers)
    
    linear_results = linear_evaluation_simple(
        model, linear_train_loader, linear_test_loader, 
        num_classes=10, epochs=50, device=device
    )
    
    # 保存所有结果
    results = {
        'config': config,
        'losses': all_losses,
        'convergence': conv_metrics,
        'feature_metrics': feature_metrics,
        'linear_results': linear_results
    }
    
    with open('logs_improved/training_results.pkl', 'wb') as f:
        pickle.dump(results, f)
    
    # ============== 绘图====================
    plt.figure(figsize=(8, 24))

    # 1. 训练损失
    plt.subplot(6, 1, 1)
    plt.plot(all_losses, label='Train Loss', color='#1f77b4')
    plt.title('Training Loss')
    plt.xlabel('Epoch')
    plt.grid(alpha=0.3)

    # 2. 收敛指标
    plt.subplot(6, 1, 2)
    plt.bar(['Rate_L', 'Delta_L', 'T_conv'], 
            [conv_metrics['loss_drop_rate'], conv_metrics['loss_fluctuation'], conv_metrics['convergence_epoch']],
            color=['#ff7f0e','#2ca02c','#d62728'])
    plt.title('Convergence Metrics')
    plt.grid(alpha=0.3)

    # 3. Train Acc
    plt.subplot(6, 1, 3)
    plt.plot(linear_results['train_acc_history'], label='Train Acc', color='#2ca02c', linewidth=2)
    plt.title('Train Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.grid(alpha=0.3)

    # 4. Test Acc
    plt.subplot(6, 1, 4)
    plt.plot(linear_results['test_acc_history'], label='Test Acc', color='#d62728', linewidth=2)
    plt.title('Test Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.grid(alpha=0.3)

    # 5. Train + Test
    plt.subplot(6, 1, 5)
    plt.plot(linear_results['train_acc_history'], label='Train Acc', color='#2ca02c')
    plt.plot(linear_results['test_acc_history'], label='Test Acc', color='#d62728')
    plt.title('Train & Test Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.grid(alpha=0.3)

    # 6. Precision / Recall / F1
    plt.subplot(6, 1, 6)
    plt.plot(linear_results['f1_history'], label='F1', linewidth=2)
    plt.plot(linear_results['precision_history'], label='Precision', linewidth=2)
    plt.plot(linear_results['recall_history'], label='Recall', linewidth=2)
    plt.title('Precision / Recall / F1')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig('logs_improved/vertical_metrics.png', dpi=300)
    plt.show()
    
    # 最终报告
    print("\n" + "="*60)
    print("FINAL FULL REPORT")
    print("="*60)
    print(f"损失下降速率 Rate_L: {conv_metrics['loss_drop_rate']:.6f}")
    print(f"损失波动幅度 ΔL: {conv_metrics['loss_fluctuation']:.4f}")
    print(f"收敛迭代数 T_conv: {conv_metrics['convergence_epoch']}")
    print(f"轮廓系数: {feature_metrics['silhouette_score']:.4f}")
    print(f"互信息 NMI: {feature_metrics['nmi_score']:.4f}")
    print(f"线性评估最佳准确率: {linear_results['best_test_acc']:.2f}%")
    print(f"F1: {linear_results['final_f1_score']:.4f}")
    print(f"精确率: {linear_results['final_precision']:.4f}")
    print(f"召回率: {linear_results['final_recall']:.4f}")
    print("="*60)
    
    return results


if __name__ == "__main__":
    torch.manual_seed(42)
    torch.cuda.manual_seed_all(42)
    np.random.seed(42)
    
    os.makedirs('checkpoints_improved', exist_ok=True)
    os.makedirs('logs_improved', exist_ok=True)
    
    try:
        results = main_improved()
        print("\nTraining completed successfully!")
    except Exception as e:
        print(f"Error: {e}")
        import traceback
        traceback.print_exc()
```
