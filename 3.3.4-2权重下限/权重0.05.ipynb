```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, normalized_mutual_info_score, f1_score, top_k_accuracy_score
import os
from tqdm import tqdm
import pickle
import sys
import matplotlib.pyplot as plt

class ResNetBackbone(nn.Module):
    """封装ResNet，用于获取特征向量"""
    def __init__(self, base_encoder, projection_dim=128):
        super().__init__()
        
        # 兼容不同PyTorch版本的权重加载方式
        try:
            # 新版本PyTorch (>=1.10) 使用 weights 参数
            self.backbone = base_encoder(weights=torchvision.models.ResNet50_Weights.DEFAULT)
        except AttributeError:
            # 旧版本PyTorch (<1.10) 使用 pretrained 参数
            self.backbone = base_encoder(pretrained=True)
        
        # 修改最后一层，使其输出特征而不是分类
        self.backbone.fc = nn.Identity()
        # 特征维度 (ResNet50 输出 2048)
        self.feature_dim = 2048  # 直接指定为2048，避免计算错误

    def forward(self, x):
        return self.backbone(x)

class ProjectionHead(nn.Module):
    """投影头: MLP, 用于将特征投影到对比学习空间"""
    def __init__(self, input_dim=2048, hidden_dim=2048, output_dim=128):
        super().__init__()
        self.mlp = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, output_dim)
        )

    def forward(self, x):
        return self.mlp(x)

class SimCLRModel(nn.Module):
    def __init__(self, backbone, projection_head):
        super().__init__()
        self.encoder = backbone
        self.projector = projection_head

    def forward(self, x):
        features = self.encoder(x)
        projections = self.projector(features)
        # L2 归一化
        projections = nn.functional.normalize(projections, dim=1)
        return projections, features  # 返回特征用于聚类

# --- 修复的数据增强 ---
class SimCLRTransform:
    """
    SimCLR 数据增强策略
    修复：处理PIL图像和Tensor两种输入
    """
    def __init__(self, image_size):
        color_jitter = transforms.ColorJitter(0.8, 0.8, 0.8, 0.2)
        
        # 主增强管道，处理PIL图像
        self.pil_train_transform = transforms.Compose([
            transforms.RandomResizedCrop(size=image_size),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomApply([color_jitter], p=0.8),
            transforms.RandomGrayscale(p=0.2),
        ])
        
        # 最终转换为Tensor和归一化
        self.to_tensor_normalize = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                std=[0.229, 0.224, 0.225]),
        ])
        
        # 用于处理张量的版本（当输入已经是张量时）
        self.tensor_train_transform = transforms.Compose([
            transforms.RandomResizedCrop(size=image_size),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomApply([transforms.ColorJitter(0.8, 0.8, 0.8, 0.2)], p=0.8),
            transforms.RandomGrayscale(p=0.2),
        ])
        
        self.image_size = image_size

    def __call__(self, x):
        # 检查输入类型
        if isinstance(x, torch.Tensor):
            # 如果输入已经是张量，先转换为PIL图像
            x = transforms.ToPILImage()(x)
        
        # 应用增强两次，得到两个不同的视图
        v1 = self.pil_train_transform(x)
        v2 = self.pil_train_transform(x)
        
        # 转换为张量并归一化
        v1 = self.to_tensor_normalize(v1)
        v2 = self.to_tensor_normalize(v2)
        
        return v1, v2

# --- 获取特征 ---
def extract_features_for_clustering(model, dataloader, device='cuda'):
    """提取所有训练数据的特征"""
    model.eval()
    features_list = []
    with torch.no_grad():
        for images, _ in tqdm(dataloader, desc="Extracting features"):
            images = images.to(device)
            _, features = model(images)  # 获取未投影的特征
            features_list.append(features.cpu())
    all_features = torch.cat(features_list, dim=0).numpy()
    return all_features

# --- 执行 ---
def perform_clustering(features, n_clusters=10):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(features)
    return cluster_labels

# --- 修复的自定义 Dataset ---
class ClusteredDataset(torch.utils.data.Dataset):
    def __init__(self, dataset, cluster_labels, transform=None):
        """
        参数:
        - dataset: 原始数据集，应该返回(PIL图像, 标签)或(图像, 标签)
        - cluster_labels: 列表
        - transform: 数据增强转换函数
        """
        self.dataset = dataset
        self.cluster_labels = cluster_labels
        self.transform = transform
        
        # 检查数据集的返回类型
        if len(self.dataset) > 0:
            sample_img, _ = self.dataset[0]
            self.input_is_pil = isinstance(sample_img, torch.Tensor)
        else:
            self.input_is_pil = False

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        img, _ = self.dataset[idx]  # 忽略原始标签
        
        if self.transform is not None:
            v1, v2 = self.transform(img)
        else:
            # 如果没有转换，使用相同的图像
            v1 = v2 = img
            
        cluster_label = self.cluster_labels[idx]
        return v1, v2, cluster_label, idx

def clustered_negative_simclr_loss(z1, z2, cluster_labels_batch, cluster_to_indices_map, temperature=0.5):
    """
    基于聚类标签选择负样本的 SimCLR 损失函数
    修正：有效负样本余弦相似度 > 0.95 时截断为 0.95，不影响训练
    """
    device = z1.device
    batch_size = z1.shape[0]

    # 1. 计算 Anchor (z1) 与 Positive (z2) 的相似度
    pos_sim = torch.sum(z1 * z2, dim=1) / temperature  # Shape: [B]

    # 2. 构建完整的特征矩阵
    z_full = torch.cat((z1, z2), dim=0)  # Shape: [2B, D]

    # ====================== 核心修复：先算原始余弦相似度，截断，再除以温度 ======================
    # 原始余弦相似度 (-1 ~ 1)
    cos_sim_full = torch.mm(z_full, z_full.T)
    
    # 3. 创建正样本掩码
    labels = torch.arange(batch_size, device=device)
    pos_mask = torch.zeros((2 * batch_size, 2 * batch_size), dtype=torch.bool, device=device)
    pos_mask[torch.arange(batch_size), torch.arange(batch_size) + batch_size] = True
    pos_mask[torch.arange(batch_size) + batch_size, torch.arange(batch_size)] = True

    # 4. 创建负样本掩码
    mask_neg = torch.ones((2 * batch_size, 2 * batch_size), dtype=torch.bool, device=device)
    mask_neg.fill_diagonal_(False)  # 排除自相似性
    mask_neg[pos_mask] = False       # 排除正样本对

    cl_labels_expanded = cluster_labels_batch.repeat(2)
    same_cluster_mask_full = (cl_labels_expanded.unsqueeze(0) == cl_labels_expanded.unsqueeze(1))
    mask_neg &= ~same_cluster_mask_full

    # 截断余弦相似度 >0.95 → 0.95（你要的 min 逻辑）
    cos_sim_full[mask_neg] = torch.clamp(cos_sim_full[mask_neg], max=0.95)

    # 最后再除以温度
    sim_matrix_full = cos_sim_full / temperature
    # ========================================================================================

    # 6. 计算损失（完全沿用你原来的逻辑）
    sim_matrix_masked = sim_matrix_full.masked_fill(~mask_neg, float('-inf'))

    # 扩展正样本相似度
    pos_sim_full = torch.cat([pos_sim, pos_sim], dim=0)

    # 计算分母：所有有效负样本的exp和
    exp_sim_neg = torch.exp(sim_matrix_masked)
    sum_exp_neg = exp_sim_neg.sum(dim=1)

    # 计算正样本的exp
    exp_pos_sim = torch.exp(pos_sim_full)

    # 计算log概率
    log_prob_pos = pos_sim_full - torch.log(exp_pos_sim + sum_exp_neg)
    loss = -log_prob_pos.mean()

    return loss

# --- 损失曲线绘制函数 ---
def plot_loss_curves(all_losses, save_path='loss_curves.png'):
    """绘制训练损失曲线"""
    plt.figure(figsize=(15, 5))
    
    # 1. 训练损失曲线
    plt.subplot(1, 3, 1)
    epochs = range(1, len(all_losses) + 1)
    plt.plot(epochs, all_losses, 'b-', linewidth=2, marker='o', markersize=4, label='Training Loss')
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Loss', fontsize=12)
    plt.title('Training Loss Curve', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend()
    
    # 2. 损失下降速率
    plt.subplot(1, 3, 2)
    if len(all_losses) > 1:
        loss_rates = []
        for i in range(1, len(all_losses)):
            rate = (all_losses[i-1] - all_losses[i])  # 每个epoch的下降量
            loss_rates.append(rate)
        
        epochs_rates = range(2, len(all_losses) + 1)
        plt.plot(epochs_rates, loss_rates, 'r-', linewidth=2, marker='s', markersize=4, label='Loss Decline Rate')
        plt.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
        plt.xlabel('Epoch', fontsize=12)
        plt.ylabel('Loss Rate', fontsize=12)
        plt.title('Loss Decline Rate per Epoch', fontsize=14)
        plt.grid(True, alpha=0.3)
        plt.legend()
    else:
        plt.text(0.5, 0.5, 'Insufficient data\nfor loss rate calculation',
                 ha='center', va='center', fontsize=12, transform=plt.gca().transAxes)
        plt.axis('off')
    
    # 3. 损失波动幅度（滑动窗口）
    plt.subplot(1, 3, 3)
    if len(all_losses) > 5:
        window_size = min(5, len(all_losses) // 2)
        loss_variations = []
        for i in range(window_size, len(all_losses)):
            window_losses = all_losses[i-window_size:i]
            variation = (max(window_losses) - min(window_losses)) / np.mean(window_losses)
            loss_variations.append(variation)
        
        epochs_variations = range(window_size + 1, len(all_losses) + 1)
        plt.plot(epochs_variations, loss_variations, 'g-', linewidth=2, marker='^', markersize=4, label='Loss Variation')
        plt.axhline(y=0.01, color='red', linestyle='--', alpha=0.7, label='1% Threshold')
        plt.xlabel('Epoch', fontsize=12)
        plt.ylabel('Loss Variation (ΔL)', fontsize=12)
        plt.title('Loss Variation (Sliding Window)', fontsize=14)
        plt.grid(True, alpha=0.3)
        plt.legend()
    else:
        plt.text(0.5, 0.5, 'Insufficient data\nfor variation calculation',
                 ha='center', va='center', fontsize=12, transform=plt.gca().transAxes)
        plt.axis('off')
    
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Loss curves saved to {save_path}")

# --- 主训练循环 ---
def main():
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")

    # 超参数
    batch_size = 256
    learning_rate = 1e-4
    temperature = 0.5
    epochs_pretrain = 1  # 仅用1个epoch获取特征进行聚类
    epochs_finetune = 100  # 主训练循环的epoch数（为演示减少）
    num_clusters = 10
    dataset_name = "CIFAR10"
    
    # 用于存储训练过程中的指标
    all_losses = []  # 每个epoch的平均损失
    epoch_losses = []  # 每个epoch中每个batch的损失（用于计算波动幅度）

    # 数据加载
    if dataset_name == "CIFAR10":
        image_size = 32
        
        #转换
        cluster_transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ])
        
        # 用于聚类的数据集
        train_dataset_raw_for_clustering = datasets.CIFAR10(
            root='./data', train=True, download=True, 
            transform=cluster_transform
        )

        # 使用临时模型进行特征提取
        temp_base_encoder = torchvision.models.resnet50
        temp_backbone = ResNetBackbone(temp_base_encoder, projection_dim=128)
        temp_projection_head = ProjectionHead(input_dim=temp_backbone.feature_dim, hidden_dim=2048, output_dim=128)
        temp_model = SimCLRModel(temp_backbone, temp_projection_head).to(device)

        cluster_loader = DataLoader(train_dataset_raw_for_clustering, batch_size=batch_size, shuffle=False, num_workers=2)
        features_for_clustering = extract_features_for_clustering(temp_model, cluster_loader, device)
        cluster_labels_all_data = perform_clustering(features_for_clustering, n_clusters=num_clusters)

        # 保存聚类标签以便重现
        cluster_labels_path = f'cluster_labels_{dataset_name}_k{num_clusters}.pkl'
        with open(cluster_labels_path, 'wb') as f:
            pickle.dump(cluster_labels_all_data, f)
        print(f"Cluster labels saved to {cluster_labels_path}")

        train_dataset_raw = datasets.CIFAR10(
            root='./data', train=True, download=True, 
            transform=None  # 不应用任何转换，返回PIL图像
        )
        
        # 创建SimCLR数据增强
        simclr_transform = SimCLRTransform(image_size=image_size)
        
        # 创建带有聚类标签的数据集
        train_dataset_with_clusters = ClusteredDataset(
            train_dataset_raw, cluster_labels_all_data, simclr_transform
        )
        
        train_loader = DataLoader(train_dataset_with_clusters, batch_size=batch_size, 
                                  shuffle=True, num_workers=2, drop_last=True)

        # 初始化主模型
        base_encoder = torchvision.models.resnet50
        backbone = ResNetBackbone(base_encoder, projection_dim=128)
        projection_head = ProjectionHead(input_dim=backbone.feature_dim, hidden_dim=2048, output_dim=128)
        model = SimCLRModel(backbone, projection_head).to(device)

        optimizer = optim.AdamW(model.parameters(), lr=learning_rate, weight_decay=1e-6)

        # 创建从聚类ID到样本索引的映射
        cluster_to_indices_map = {}
        for idx, label in enumerate(cluster_labels_all_data):
            if label not in cluster_to_indices_map:
                cluster_to_indices_map[label] = []
            cluster_to_indices_map[label].append(idx)

        # 训练循环
        model.train()
        for epoch in range(epochs_finetune):
            total_loss = 0
            num_batches = 0
            batch_losses = []  # 存储当前epoch中每个batch的损失

            pbar = tqdm(train_loader, desc=f"Main Epoch {epoch+1}/{epochs_finetune}")
            for v1, v2, cluster_labels_batch, indices_batch in pbar:
                v1, v2 = v1.to(device), v2.to(device)
                cluster_labels_batch = cluster_labels_batch.to(device)  # [batch_size]

                optimizer.zero_grad()

                # 前向传播
                z1, _ = model(v1)  # z1, z2 是投影特征
                z2, _ = model(v2)

                # 计算带聚类负样本的损失
                loss = clustered_negative_simclr_loss(z1, z2, cluster_labels_batch, 
                                                       cluster_to_indices_map, temperature=temperature)

                loss.backward()
                optimizer.step()

                batch_loss = loss.item()
                total_loss += batch_loss
                num_batches += 1
                batch_losses.append(batch_loss)

                pbar.set_postfix({'Batch Loss': batch_loss})

            avg_loss = total_loss / num_batches
            all_losses.append(avg_loss)
            epoch_losses.append(batch_losses)  # 保存每个batch的损失
            
            print(f'Main Epoch [{epoch+1}/{epochs_finetune}], Average Loss: {avg_loss:.4f}')

        # 保存训练好的编码器
        torch.save(model.encoder.state_dict(), f'clustered_simclr_encoder_{dataset_name}.pth')
        print("Pre-trained encoder saved.")

        # --- 计算损失收敛性指标 ---
        print("\n=== Loss Convergence Metrics ===")
        
        # 1. 损失下降速率: Rate_L = (L_0 - L_T) / T
        if len(all_losses) >= 2:
            L_0 = all_losses[0]
            L_T = all_losses[-1]
            T = len(all_losses) - 1  # 从epoch 0到T
            rate_L = (L_0 - L_T) / T if T > 0 else 0
            print(f"损失下降速率 (Rate_L): {rate_L:.6f}")
        
        # 2. 损失波动幅度: ΔL = (max(L_t) - min(L_t)) / mean(L_t), t ∈ [T/2, T]
        if len(all_losses) >= 2:
            half_idx = len(all_losses) // 2
            if half_idx < len(all_losses):
                late_losses = all_losses[half_idx:]
                max_loss = max(late_losses)
                min_loss = min(late_losses)
                mean_loss = np.mean(late_losses)
                if mean_loss > 0:
                    delta_L = (max_loss - min_loss) / mean_loss
                    print(f"损失波动幅度 (ΔL): {delta_L:.6f}")
                else:
                    print(f"损失波动幅度 (ΔL): N/A (mean loss is zero)")
        
        # 3. 收敛迭代数 T_conv (损失波动幅度 < 1% 时的迭代数)
        if len(epoch_losses) > 0:
            # 使用滑动窗口检测稳定性
            window_size = min(5, len(epoch_losses) // 2)
            T_conv = None
            for i in range(window_size, len(all_losses)):
                window_losses = all_losses[i-window_size:i]
                max_loss = max(window_losses)
                min_loss = min(window_losses)
                mean_loss = np.mean(window_losses)
                if mean_loss > 0:
                    variation = (max_loss - min_loss) / mean_loss
                    if variation < 0.01:  # 1%阈值
                        T_conv = i
                        break
            
            if T_conv is not None:
                print(f"收敛迭代数 (T_conv): {T_conv} epochs")
            else:
                print(f"收敛迭代数 (T_conv): Not converged within {len(all_losses)} epochs")
        
        # 绘制Loss曲线
        plot_loss_curves(all_losses, save_path='loss_curves.png')
        
        # --- 特征质量指标计算 ---
        print("\n=== Feature Quality Metrics ===")
        
        # 重新加载原始训练集（带标签）用于特征质量评估
        train_dataset_for_features = datasets.CIFAR10(
            root='./data', train=True, download=True,
            transform=cluster_transform
        )
        
        feature_loader = DataLoader(train_dataset_for_features, batch_size=batch_size, shuffle=False, num_workers=2)
        
        # 提取特征
        model.eval()
        all_features = []
        all_true_labels = []
        
        with torch.no_grad():
            for images, labels in tqdm(feature_loader, desc="Extracting features for quality evaluation"):
                images = images.to(device)
                _, features = model(images)
                all_features.append(features.cpu().numpy())
                all_true_labels.append(labels.numpy())
        
        all_features = np.vstack(all_features)
        all_true_labels = np.concatenate(all_true_labels)
        
        # 1. 轮廓系数 (Silhouette Coefficient)
        print("Calculating Silhouette Coefficient...")
        # 由于计算轮廓系数需要大量内存和时间，我们采样计算
        sample_size = min(5000, len(all_features))
        indices = np.random.choice(len(all_features), sample_size, replace=False)
        
        # 使用KMeans聚类
        kmeans = KMeans(n_clusters=10, random_state=42, n_init=5)
        cluster_labels = kmeans.fit_predict(all_features[indices])
        
        silhouette = silhouette_score(all_features[indices], cluster_labels)
        print(f"轮廓系数 (Silhouette Coefficient): {silhouette:.4f}")
        
        # 2. 互信息得分 (Mutual Information Score)
        print("Calculating Mutual Information Score...")
        nmi = normalized_mutual_info_score(all_true_labels[indices], cluster_labels)
        print(f"互信息得分 (Normalized Mutual Information): {nmi:.4f}")
        
        # --- 线性评估 ---
        print("\n=== Downstream Task Evaluation ===")
        
        # 线性评估的转换
        linear_eval_transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ])
        
        # 加载数据集
        linear_train_dataset = datasets.CIFAR10(root='./data', train=True, 
                                                transform=linear_eval_transform, download=True)
        linear_test_dataset = datasets.CIFAR10(root='./data', train=False, 
                                               transform=linear_eval_transform, download=True)
        
        linear_train_loader = DataLoader(linear_train_dataset, batch_size=batch_size, 
                                          shuffle=True, num_workers=2)
        linear_test_loader = DataLoader(linear_test_dataset, batch_size=batch_size, 
                                         shuffle=False, num_workers=2)

        def evaluate_linear(model, train_loader, test_loader, num_classes=10, epochs=20, device='cuda'):
            """线性评估函数，返回详细指标"""
            model.eval()
            for param in model.parameters():
                param.requires_grad = False

            linear_classifier = nn.Linear(model.encoder.feature_dim, num_classes).to(device)
            optimizer_cls = optim.Adam(linear_classifier.parameters(), lr=1e-3)
            criterion_cls = nn.CrossEntropyLoss()
            
            # 存储训练过程中的指标
            train_acc_history = []
            test_acc_history = []
            train_f1_history = []
            
            # 训练线性分类器
            for epoch in range(epochs):
                linear_classifier.train()
                total_loss = 0
                correct = 0
                total = 0
                
                all_preds = []
                all_labels = []
                
                for images, labels in tqdm(train_loader, desc=f"Linear Eval Epoch {epoch+1}/{epochs}"):
                    images, labels = images.to(device), labels.to(device)
                    optimizer_cls.zero_grad()

                    with torch.no_grad():
                        features = model.encoder(images)

                    logits = linear_classifier(features)
                    loss = criterion_cls(logits, labels)
                    loss.backward()
                    optimizer_cls.step()

                    total_loss += loss.item()
                    _, predicted = logits.max(1)
                    total += labels.size(0)
                    correct += predicted.eq(labels).sum().item()
                    
                    all_preds.extend(predicted.cpu().numpy())
                    all_labels.extend(labels.cpu().numpy())

                train_acc = 100. * correct / total
                train_acc_history.append(train_acc)
                
                # 计算F1分数
                f1 = f1_score(all_labels, all_preds, average='macro')
                train_f1_history.append(f1)
                
                # 在测试集上评估
                linear_classifier.eval()
                test_correct = 0
                test_total = 0
                test_preds = []
                test_true = []
                
                with torch.no_grad():
                    for images, labels in test_loader:
                        images, labels = images.to(device), labels.to(device)
                        features = model.encoder(images)
                        logits = linear_classifier(features)
                        _, predicted = logits.max(1)
                        test_total += labels.size(0)
                        test_correct += predicted.eq(labels).sum().item()
                        
                        test_preds.extend(predicted.cpu().numpy())
                        test_true.extend(labels.cpu().numpy())
                
                test_acc = 100. * test_correct / test_total
                test_acc_history.append(test_acc)
                
                print(f'Linear Eval Epoch [{epoch+1}/{epochs}], Train Acc: {train_acc:.2f}%, Test Acc: {test_acc:.2f}%, F1: {f1:.4f}')
            
            # 最终测试集评估（计算所有指标）
            linear_classifier.eval()
            test_correct = 0
            test_total = 0
            all_test_preds = []
            all_test_true = []
            all_test_probs = []
            
            with torch.no_grad():
                for images, labels in test_loader:
                    images, labels = images.to(device), labels.to(device)
                    features = model.encoder(images)
                    logits = linear_classifier(features)
                    probs = torch.softmax(logits, dim=1)
                    
                    _, predicted = logits.max(1)
                    test_total += labels.size(0)
                    test_correct += predicted.eq(labels).sum().item()
                    
                    all_test_preds.extend(predicted.cpu().numpy())
                    all_test_true.extend(labels.cpu().numpy())
                    all_test_probs.extend(probs.cpu().numpy())

            # 计算准确率
            accuracy = 100. * test_correct / test_total
            
            # 计算F1分数
            final_f1 = f1_score(all_test_true, all_test_preds, average='macro')
            
            # 计算Top-k准确率
            all_test_probs = np.array(all_test_probs)
            all_test_true = np.array(all_test_true)
            
            # Top-1准确率（与准确率相同）
            top1_accuracy = accuracy
            
            # Top-3准确率
            top3_accuracy = top_k_accuracy_score(all_test_true, all_test_probs, k=3) * 100
            
            # Top-5准确率
            top5_accuracy = top_k_accuracy_score(all_test_true, all_test_probs, k=5) * 100
            
            return {
                'accuracy': accuracy,
                'top1_accuracy': top1_accuracy,
                'top3_accuracy': top3_accuracy,
                'top5_accuracy': top5_accuracy,
                'f1_score': final_f1,
                'train_acc_history': train_acc_history,
                'test_acc_history': test_acc_history,
                'train_f1_history': train_f1_history
            }
        
        # 执行线性评估
        eval_results = evaluate_linear(model, linear_train_loader, linear_test_loader, 
                                       num_classes=10, epochs=20, device=device)
        
        # 输出下游任务性能指标
        print("\n=== Downstream Task Performance Metrics ===")
        print(f"线性评估准确率: {eval_results['accuracy']:.2f}%")
        print(f"Top-1准确率: {eval_results['top1_accuracy']:.2f}%")
        print(f"Top-3准确率: {eval_results['top3_accuracy']:.2f}%")
        print(f"Top-5准确率: {eval_results['top5_accuracy']:.2f}%")
        print(f"F1分数: {eval_results['f1_score']:.4f}")
        
        # --- 绘制线性评估结果曲线 ---
        print("\n=== Generating Linear Evaluation Plots ===")
        
        # 1. 训练和测试准确率曲线
        plt.figure(figsize=(12, 5))
        
        plt.subplot(1, 2, 1)
        epochs = range(1, len(eval_results['train_acc_history']) + 1)
        plt.plot(epochs, eval_results['train_acc_history'], 'b-', linewidth=2, marker='o', markersize=4, label='Train Accuracy')
        plt.plot(epochs, eval_results['test_acc_history'], 'r-', linewidth=2, marker='s', markersize=4, label='Test Accuracy')
        plt.xlabel('Epoch', fontsize=12)
        plt.ylabel('Accuracy (%)', fontsize=12)
        plt.title('Linear Evaluation Accuracy', fontsize=14)
        plt.grid(True, alpha=0.3)
        plt.legend()
        
        # 2. F1分数曲线
        plt.subplot(1, 2, 2)
        plt.plot(epochs, eval_results['train_f1_history'], 'g-', linewidth=2, marker='^', markersize=4, label='F1 Score')
        plt.xlabel('Epoch', fontsize=12)
        plt.ylabel('F1 Score', fontsize=12)
        plt.title('F1 Score Progression', fontsize=14)
        plt.grid(True, alpha=0.3)
        plt.legend()
        
        plt.tight_layout()
        plt.savefig('linear_evaluation_curves.png', dpi=300, bbox_inches='tight')
        plt.show()
        print("Linear evaluation curves saved to linear_evaluation_curves.png")
        
        # --- 指标总结 ---
        print("\n" + "="*60)
        print("SUMMARY OF ALL EVALUATION METRICS")
        print("="*60)
        
        print("\n1. Loss Convergence Metrics:")
        if len(all_losses) >= 2:
            print(f"   Loss下降速率 (Rate_L): {rate_L:.6f}")
            if 'delta_L' in locals():
                print(f"   损失波动幅度 (ΔL): {delta_L:.6f}")
            if T_conv is not None:
                print(f"   收敛迭代数 (T_conv): {T_conv} epochs")
        
        print("\n2. Feature Quality Metrics:")
        print(f"   轮廓系数 (Silhouette Coefficient): {silhouette:.4f}")
        print(f"   互信息得分 (Normalized Mutual Information): {nmi:.4f}")
        
        print("\n3. Downstream Task Performance Metrics:")
        print(f"   线性评估准确率: {eval_results['accuracy']:.2f}%")
        print(f"   Top-1准确率: {eval_results['top1_accuracy']:.2f}%")
        print(f"   Top-3准确率: {eval_results['top3_accuracy']:.2f}%")
        print(f"   Top-5准确率: {eval_results['top5_accuracy']:.2f}%")
        print(f"   F1分数: {eval_results['f1_score']:.4f}")
        
        print("\n4. Generated Plots:")
        print("   - loss_curves.png: 训练损失曲线")
        print("   - linear_evaluation_curves.png: 线性评估曲线")
        print("="*60)

if __name__ == "__main__":
    # 设置matplotlib中文字体（如果系统支持）
    try:
        plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
        plt.rcParams['axes.unicode_minus'] = False
        print("Chinese font support enabled for plots.")
    except:
        print("Using default matplotlib fonts.")
    
    main()
```
